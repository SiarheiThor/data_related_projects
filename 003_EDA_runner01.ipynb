```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.dtypes
```

```python
df['training_data']
```

```python
# Define a function that converts a string to a list
def to_list(s):
    return s.strip('[]').split(', ')

# Convert the training_data column back to lists
df['training_data'] = df['training_data'].apply(to_list)

df.head()
```

```python
# Create a function to split the training_data column into three separate columns
def split_training_data(row):
    distance, time, pace = row['training_data']
    return pd.Series({'distance': distance, 'time': time, 'pace': pace})

# Apply the function to the dataframe
df[['distance', 'time', 'pace']] = df.apply(split_training_data, axis=1)

# Print the dataframe
print(df.head())
```

```python
df.head()
```

```python
# Define a function that removes letters and spaces
def remove_letters_and_spaces(s):
    return s.replace(' km', '').replace(' ', '').replace('/km', '').replace("'", "")

# Remove letters and spaces from the distance, time, and pace columns
df['distance'] = df['distance'].apply(remove_letters_and_spaces)
df['time'] = df['time'].apply(remove_letters_and_spaces)
df['pace'] = df['pace'].apply(remove_letters_and_spaces)
```

```python
df.head()
```

```python
df_cleaned = df.drop(columns=['training_data'])
```

```python
df_cleaned.head(20)
```

```python
replace_dict = {'Idag':'january 02, 2023', 'I går':'january 01, 2023', 'Lördag':'december 31, 2022', 'Fredag':'december 30, 2022', 'Torsdag':'december 29, 2022', 'Onsdag':'december 28, 2022', 'Tisdag':'december 27, 2022'}
df_cleaned['date'] = df_cleaned['date'].replace(replace_dict)
df_cleaned.head(20)
```

```python
df_cleaned
```

```python

month_mapping = {
'january': 'January',
'februari': 'February',
'mars': 'March',
'april': 'April',
'maj': 'May',
'juni': 'June',
'juli': 'July',
'augusti': 'August',
'september': 'September',
'oktober': 'October',
'november': 'November',
'december': 'December'
}

df_cleaned['date'] = df_cleaned['date'].map(lambda date: f"{month_mapping[date.split()[0]]} {' '.join(date.split()[1:])}")
```

```python
df_cleaned.head(20)
```

```python
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'])
```

```python
df_cleaned['distance'] = df_cleaned['distance'].astype(float)
```

```python
df_cleaned.head()
```

```python
import datetime

def transform_time(time_str):
    # Check if time_str is in the hh:mm:ss format
    try:
        # Convert time_str to a timedelta object
        return pd.to_timedelta(time_str)
    except ValueError:
        # If time_str is not in the hh:mm:ss format, it must be in the mm:ss or mm:ss.ss format
        # Split time_str into minutes and seconds
        minutes, seconds = time_str.split(':')
        # Convert minutes and seconds to seconds
        total_seconds = int(minutes) * 60 + float(seconds)
        # Convert total_seconds to a timedelta object and return it
        return pd.to_timedelta(total_seconds, unit='s')

df_cleaned['time'] = df_cleaned['time'].apply(transform_time)
df_cleaned['pace'] = df_cleaned['pace'].apply(transform_time)
```

```python
df_cleaned.head()
```

```python
df_cleaned['time'] = pd.to_timedelta(df_cleaned['time'])
df_cleaned['pace'] = pd.to_timedelta(df_cleaned['pace'])
df_cleaned.head()
```
